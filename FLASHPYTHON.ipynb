```python
class Fractal:
    def __init__(self,app):
        self.app = app
        self.screen_array = np.full((width,height,3),[0,0,0],dtype=np.uint8)
#         self.x = np.linspace(0,width,num=width,dtype=np.uint8)
#         self.y = np.linspace(0,height,num=height,dtype=np.uint8)
        self.vel = 0.01
        self.scale = 0.993
        self.increment = np.array([0.0,0.0])
        self.max_iter= 500
        self.max_iter_limit = 5000
        self.zoom = 2.2/height
        offset = np.array([width*0.3,height])//2

        self.texture = pygame.image.load("C:\\Users\\SHIVA SINGH\\Downloads\\pawel-czerwinski-ruJm3dBXCqw-unsplash.png")
        self.texture_size = min(texture.get_size())-1
        self.texture_array = pygame.surfarray.array3d(texture)

        #delta_time
        self.app_speed = 1/3000
        self.prev_time = pygame.time.get_ticks()
    
    
    def delta_time(self):
        time_now = pygame.time.get_ticks()-self.prev_time
        self.prev_time = time_now
        return time_now*self.app_speed
    
    @staticmethod
    #@cuda.jit
    @numba.njit(nopython=True,fastmath=True,parallel=True)
    def render(screen_array,texture_array,texture_size,max_iter,zoom,dx,dy):
        for x in numba.prange(width):
            for y in range(height):
                c = complex((x-offset[0])*zoom-dx ,(y-offset[1])*zoom-dy)
                z = 0
                num_iter =0
                for i in range(max_iter):
                    z = z**2+c
                    if(z.real**2+z.imag**2>4):
                        break
                    num_iter+=1
                col = int(texture_size*num_iter/max_iter)
                screen_array[x,y] = texture_array[col,col]
        return screen_array
            
#         x = (self.x-offset[0])*zoom
#         y = (self.y-offset[1])*zoom
#         c = x+1j*y[:,None]
        
#         num_iter = np.full(c.shape,max_iter)
        
#         z = np.empty(c.shape,np.complex64)
        
#         for i in range(max_iter):
#             mask = (num_iter==max_iter)
#             z[mask] = z[mask]**2 + c[mask]
#             num_iter[mask&(np.abs(z)>2.0)] = i+1
        
#         col = (num_iter.T*texture_size/max_iter).astype(np.uint8)
#         self.screen_array = texture_array[col,col]
    
    def control(self):
        pressed_key = pygame.key.get_pressed()
        dt = self.delta_time()
        if pressed_key[pygame.K_a]:
            self.increment[0] += self.vel * dt
        if pressed_key[pygame.K_d]:
            self.increment[0] -= self.vel * dt
        if pressed_key[pygame.K_w]:
            self.increment[1] += self.vel * dt
        if pressed_key[pygame.K_s]:
            self.increment[1] -= self.vel * dt
        
        #zoom
        
        if pressed_key[pygame.K_UP]:
            self.zoom*=self.scale
            self.vel*=self.scale
        if pressed_key[pygame.K_DOWN]:
            inv_scale = 2-self.scale
            self.zoom*=inv_scale
            self.vel*=inv_scale
        
        if pressed_key[pygame.K_LEFT]:
            self.max_iter -= 1
            #print(max_iter)
        if pressed_key[pygame.K_RIGHT]:
            self.max_iter+=1
            #print(max_iter)
        self.max_iter = min(max(self.max_iter,2),self.max_iter_limit)
    
    def update(self):
        self.control()
        self.screen_array = self.render(self.screen_array,self.texture_array,self.texture_size,self.max_iter,self.zoom,self.increment[0],self.increment[1])
        
    
    def draw(self):
        icon = pygame.image.load("C:\\Users\\SHIVA SINGH\Downloads\\lightning.png")
        pygame.display.set_icon(icon)
        pygame.surfarray.blit_array(self.app.screen,self.screen_array)
    
    def run(self):
        self.update()
        self.draw()
        
```

```python
import os
```

```python
class App:
    def __init__(self):
        self.screen = pygame.display.set_mode(res)
        self.clock = pygame.time.Clock()
        self.fractal = Fractal(self)

    def run(self):
        running  = True
        frames = 0
        while running:
            for event in pygame.event.get():
                if event.type==pygame.QUIT:
                    running = False
            self.screen.fill('black')
            self.fractal.run()
            #save
            filename = 'frame{}.png'.format(frames)
            filepath = os.path.join('C:\\Users\\SHIVA SINGH\\Documents\\fractal', filename)
            pygame.image.save(self.screen, filepath)
            pygame.display.flip()
            self.clock.tick()
            pygame.display.set_caption(f"FPS:{self.clock.get_fps()}")
            frames+=1
        pygame.quit()
           

if __name__ == '__main__':
    app = App()
    app.run()
        
```
